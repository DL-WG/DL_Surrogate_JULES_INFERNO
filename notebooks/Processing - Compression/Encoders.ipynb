```python
# Function we use for plotting
def plot_data_img(nr, nc, data, true_data, ts, title='Plot test', name_file=None, threshold=0.01):
  '''
  Function to plot the maps of the variables
  nr : number of plot row - int
  nc : number of plot column - int
  true_data : the original data - numpy array
  ts :  indices of the maps to plot - list
  title : title of the maps plotted - string
  name_file : name to save the plot as images. - string
  threshold : value added to the array to avoid nan values when converting to logarithmic scale - int
  '''
  #The min and max value of the oiginal data are saved to have the same colorbar when plotting the maps. 
  valmin = [np.min(true_data[elem]) for elem in ts]
  valmax = [np.max(true_data[elem]) for elem in ts] 
  
  fig, axes = plt.subplots(nr, nc, figsize=(20, 13))
  for t, ax, vmin, vmax in zip(ts, axes.ravel(), valmin, valmax):
    #Flip the image to see in the proper orientation
    rotate_image = ndimage.rotate(np.log(data[t]+threshold), 180)
    #Show the map with the threshold added.
    im = ax.imshow(rotate_image, vmin=np.log(vmin+threshold), vmax=np.log(vmax+threshold))
    ax.set_title(title + f" {t}")
    ax.set_ylabel('Latitude')
    ax.set_xlabel('Longitude')
    #displaying the colorbar and adjusting its size
    fig.colorbar(im, ax=ax, fraction=0.03, pad=0.03)
    #Adjust maps size to avoid overlapping 
    fig.tight_layout(pad=1.5)
  
  #The image is saved only if a name_file is given to the function
  if name_file != None:
    #Adjust the size of the image to reduce white space aroung the image
    fig.set_size_inches(15,2.5)
    plt.savefig("/content/drive/MyDrive/Wildfires_IRP/Figures/"+name_file)

  plt.show()

def plot_data_img_diff(nr, nc, data1, data2, ts, title='Plot test', name_file=None):
  '''
  Function to plot the difference between original maps and maps predicted by the models
  nr : number of plot row - int
  nc : number of plot column - int
  data1 : the original data - numpy array
  data2 : predictions - numpy array
  ts :  indices of the maps to plot - list
  title : title of the maps plotted - string
  name_file : name to save the plot as images. - string
  '''
  fig, axes = plt.subplots(nr, nc, figsize=(20, 13))
  for t, ax in zip(ts, axes.ravel()):
    #Flip the image to see in the proper orientation
    rotate_image = ndimage.rotate(data2[t] - data1[t], 180)
    #Show the map with the threshold in a gray scale.
    im = ax.imshow(rotate_image, cmap='gray')
    ax.set_title(title + f" {t}")
    ax.set_ylabel('Latitude')
    ax.set_xlabel('Longitude')
    #displaying the colorbar and adjusting its size
    fig.colorbar(im, ax=ax, fraction=0.03, pad=0.03)
    #Adjust maps size to avoid overlapping 
    fig.tight_layout(pad=1.5)
  
  #The image is saved only if a name_file is given to the function
  if name_file != None:
    #Adjust the size of the image to reduce white space aroung the image
    fig.set_size_inches(15,2.5) # this is use to when save the figures to avoid white space above and under
    plt.savefig("/content/drive/MyDrive/Wildfires_IRP/Figures/"+name_file, dpi=300)

  plt.show()

def compressors_new_var(data, saving_name, show_res = 1, summary = False, epochs = 1000):
  '''
  The function built a CAE compressor for a specific variable.
  Training and testing are split and Callbacks are defined within the function.
  Funtion builds and compile the CAE.
  CAE summary, training progress and metrics, and training performances are displayed.

  data : data of  the variable to compress - numpy array
  saving_name : saving path of the trained model - string
  show_res : type of result display (0 : None, 1:epochs progess, 2:detailed progress) - int
  summary : display summary of the model (0: No, 1: Yes) - int
  epochs : maximum number of epochs the model realises - int 
  '''

  #splitting data into a train and test set
  #test size = 25% of the data
  #random_state = 123 ensures the same split each time
  #Shuffle mmixe the data when splitting. 
  train_sc_img, test_sc_img = train_test_split(data, test_size=0.25, random_state=123, shuffle=True)
  print("\nShape of training set for CAE :", train_sc_img.shape)
  print("Shape of testing set for CAE :", test_sc_img.shape)
  print("\n")

  #Early stopping callback is defined here on validation loss. 
  #The training is stop if loss doesn't change for 50 epochs.
  #Best weights encountered so far are restored when stopping
  early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss',
                                                min_delta=0.0,
                                                patience=50,
                                                verbose=2,
                                                restore_best_weights=True)

  #Defining Encoder
  conv_encoder = keras.models.Sequential([
      keras.layers.InputLayer(input_shape=(112, 192, 1)),
      keras.layers.Conv2D(128, kernel_size=10, padding="SAME", activation="relu"),
      keras.layers.LeakyReLU(alpha=0.1),
      keras.layers.MaxPool2D(),
      keras.layers.Conv2D(64, kernel_size=8, padding="SAME", activation="relu"),
      keras.layers.LeakyReLU(alpha=0.1),
      keras.layers.MaxPool2D(),
      keras.layers.Conv2D(32, kernel_size=5, padding="SAME", activation="relu"),
      keras.layers.LeakyReLU(alpha=0.1),
      keras.layers.Flatten(),
      keras.layers.Dropout(0.1),
      keras.layers.Dense(128, activation="relu"),
      keras.layers.LeakyReLU(alpha=0.1),
      keras.layers.Dropout(0.1),
      keras.layers.Dense(64, activation="relu"),
      keras.layers.LeakyReLU(alpha=0.1),
      keras.layers.Dropout(0.1),
      keras.layers.Dense(15)
  ])

  #define decoder
  conv_decoder = keras.models.Sequential([
      keras.layers.Dense(64, input_shape=[15], activation="relu"),
      keras.layers.LeakyReLU(alpha=0.1),
      keras.layers.Dropout(0.1),
      keras.layers.Dense(128, activation="relu"),
      keras.layers.LeakyReLU(alpha=0.1),
      keras.layers.Dropout(0.1),
      keras.layers.Dense(28*48*32, activation="relu"), # change it to 28, 48
      keras.layers.LeakyReLU(alpha=0.1),
      keras.layers.Reshape(target_shape=(28, 48, 32)),
      keras.layers.LeakyReLU(0.1),
      keras.layers.Conv2D(32, kernel_size=5, padding="SAME", activation="relu"),
      keras.layers.LeakyReLU(0.1),
      keras.layers.UpSampling2D(),
      keras.layers.Conv2D(64, kernel_size=8, padding="SAME", activation="relu"),
      keras.layers.LeakyReLU(0.1),
      keras.layers.UpSampling2D(),
      keras.layers.Conv2D(128, kernel_size=10, padding="SAME", activation="relu"),
      keras.layers.LeakyReLU(0.1),
      keras.layers.Conv2D(1, kernel_size=15, strides=1, padding="SAME")
  ])



  conv_ae = keras.models.Sequential([conv_encoder, conv_decoder])

  #Compiling model with adam optimizer and mse loss
  conv_ae.compile(loss="mse", optimizer='adam')
  if summary == True :
    conv_encoder.summary()
    conv_decoder.summary()
    print("\n")

  #Building the compressor
  conv_ae = keras.models.Sequential([conv_encoder, conv_decoder])
  conv_ae.compile(loss="mse", optimizer='adam')
  
  #Training the compressor
  start = time.time()
  history = conv_ae.fit(train_sc_img, train_sc_img, 
                        epochs=epochs,
                        verbose=show_res,
                        batch_size=30,
                        validation_split=0.25,
                        callbacks=[early_stopping])
  end = time.time()

  #Plot the training history - the evoluion of metrics along training
  print("\n______________ Loss plot of CAE__________________\n")
  plt.figure(figsize=(12,5))
  plt.plot(history.history['loss'])
  plt.plot(history.history['val_loss'])
  plt.legend(['Loss', 'Val_loss']);

  #display best score of the training
  print("\nBest score : ", conv_ae.evaluate(train_sc_img, train_sc_img, verbose=0))

  #Display results of the compressor on train and test set
  comp_sc_img = conv_encoder.predict(train_sc_img)
  decomp_sc_img = conv_decoder.predict(comp_sc_img)
  mse_train = np.linalg.norm(decomp_sc_img.ravel() - train_sc_img.ravel()) / np.linalg.norm(train_sc_img.ravel())
  print("\n___________Training set compression______________")
  print("Dimension before compression : ", train_sc_img.shape)
  print("Dimension after compression : ", comp_sc_img.shape)
  print("Dimension after decompression : ", decomp_sc_img.shape)
  print('mse for train dataset: ', mse_train)
  print('\n Time to train AE: ', round(end-start, 2), 's')

  comp_sc_testimg = conv_encoder.predict(test_sc_img)
  decomp_sc_testimg = conv_decoder.predict(comp_sc_testimg)
  mse_test =np.linalg.norm(decomp_sc_testimg.ravel() - test_sc_img.ravel()) / np.linalg.norm(test_sc_img.ravel())
  print("\n___________Testing set compression______________")
  print("Dimension before compression : ", test_sc_img.shape)
  print("Dimension after compression : ", comp_sc_testimg.shape)
  print("Dimension after decompression : ", decomp_sc_testimg.shape)
  print('mse for train dataset: ', mse_test)

  #save the compressor
  conv_ae.save(f'/content/drive/MyDrive/Wildfires_IRP/compressors/{saving_name}')

  return conv_ae
```

```python
#AE takes also 1D data. nFeatures is the inital shape of the vector therefore the input dimension of AE first layer.
nFeatures = np.shape(fdta_train_1D)[1]

#Early stopping callback is defined here on validation loss. 
#The training is stop if loss doesn't change for 50 epochs.
#Best weights encountered so far are restored when stopping
early_stop = EarlyStopping(monitor = 'val_loss', min_delta = 0.0, patience = 50, verbose = 2, restore_best_weights=True)

#Define my NN
encoder = Sequential()
encoder.add(Dense(128, input_shape = [nFeatures], activation = 'relu'))
encoder.add(LeakyReLU(0.2))
encoder.add(Dropout(0.2))
encoder.add(Dense(64))
encoder.add(LeakyReLU(0.2))
encoder.add(Dropout(0.2))
encoder.add(Dense(15))

decoder = Sequential()
decoder.add(Dense(64, input_shape=[15], activation = 'relu'))
decoder.add(LeakyReLU(0.2))
encoder.add(Dropout(0.2))
decoder.add(Dense(128, input_shape=[64], activation='elu'))
decoder.add(LeakyReLU(0.2))
encoder.add(Dropout(0.2))
decoder.add(Dense(nFeatures))

#Encoder model
encoder.summary()
#Decoder model
decoder.summary()
```

```python
#ae model
ae_model = Sequential([encoder, decoder])
ae_model.summary()
```

```python
ae_model.compile(loss="mse", 
                optimizer=keras.optimizers.Adam(learning_rate=0.0001, beta_1=0.9, beta_2=0.999))
```

```python
epochs = 1000
start = time.time()
history = ae_model.fit(fdta_train_1D, fdta_train_1D,
                       epochs=epochs,
                       verbose=1,
                       validation_split=0.25, 
                       callbacks=[early_stop], shuffle=True)

end = time.time()
```
