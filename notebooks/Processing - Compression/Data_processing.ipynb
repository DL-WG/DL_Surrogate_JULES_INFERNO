# Useful Functions

```python
# Defining a function to process the environmental variables netcdf datasets to a grid box numpy arrays  
def process_new_data(full_data, x_list, y_list, file, save_name):
  '''
  full_data : list of the values for a specific environmental variable
  x_list : numpy list of the available latitudes
  y_list : numpy list of the available longitudes
  file : netcdf file containing JULES-INFERNO outputs
  save_name : text as the name to save the process data as numpy array
  '''
  dta= []
  #Loop creating a grid box for each monthly data
  for mth in range(0, len(full_data)):
    #Defining a grid of 0 values of dimensions equals to latitudes and longitudes
    field = np.zeros((len(x_list),len(y_list)))

    #Completing the grid with the values of 'full_data' corressponding to the latitude and longitude.
    for i in range(np.array(file['latitude']).size):
      np.array(file['latitude']).ravel()[i]
      index_x = x_list.index(np.array(file['latitude']).ravel()[i])
      index_y = y_list.index(np.array(file['longitude']).ravel()[i])

      field[index_x,index_y] = full_data[mth].ravel()[i]

    #Adding the monthly map to the general array
    dta.append(field)
    
    #Indicator of progress
    if mth%10 == 0:
      print(f'Until month {mth} : Done')

  #Converting list to array and saving it 
  fdta_as_images = np.asarray(dta)
  np.save(f"/content/drive/MyDrive/Wildfires_IRP/data/{save_name}.npy", fdta_as_images)

#The reference netcdf file to extract the latitude and longitude 
file = netCDF4.Dataset('/content/drive/MyDrive/Wildfires_IRP/data/JULES-ES.1p0.vn5.4.50.fireMIP.LGM.365x30.p1.S2.Annual.1961.nc')
x_list = list(np.unique(np.array(file['latitude'])))
y_list = list(np.unique(np.array(file['longitude'])))
```
